```python
import pandas as pd
```

```python
toronto_collisions = pd.read_csv('Resources/Fatal_Collisions_toronto.csv')
```

```python
toronto_collisions.head()
```

```python
toronto_collisions.columns
```

```python
toronto_collisions=toronto_collisions.rename(columns={'X':'Longitude','Y':'Latitude','REDLIGHT':'Red Light','ALCOHOL':'Alcohol'})

```

```python
toronto_collision=toronto_collisions[['Longitude', 'Latitude', 'YEAR', 'DATE', 'TIME', 'Hour',
       'PEDESTRIAN', 'CYCLIST', 'AUTOMOBILE', 'MOTORCYCLE', 'TRUCK']]

```

```python
toronto_collision
```

```python
toronto_collision=toronto_collision.sort_values('YEAR',ascending=False)
```

```python
toronto_collision.dtypes
```

```python
toronto_collision['Time Patterns']=0
```

```python
#Time PATTERNS
toronto_collision['Time Patterns'][toronto_collision['Hour'].isin(range(1,5))]='Early Morning'
toronto_collision['Time Patterns'][toronto_collision['Hour'].isin(range(4,10))]='Morning Commute'
toronto_collision['Time Patterns'][toronto_collision['Hour'].isin(range(9,18))]='Working Hours'
toronto_collision['Time Patterns'][toronto_collision['Hour'].isin(range(17,21))]='Evening Commute'
toronto_collision['Time Patterns'][toronto_collision['Hour'].isin(range(20,25))]='Evening'
```

```python
toronto_collision.head()
```

```python
time_groups=toronto_collision.groupby('Time Patterns')
```

```python
years=list(toronto_collision['YEAR'].unique())
years
```

```python
# create df for each year to collect annual data 
dfs=[]
for year in years:
    year_toronto=toronto_collision.loc[toronto_collision['YEAR']==year]
    dfs.append(year_toronto)
```

```python
# Collect Dataframes for each year
year_eighteen=dfs[0]
year_seventeen=dfs[1]
year_sixteen=dfs[2]
year_fifteen=dfs[3]
year_fourteen=dfs[4]
year_thirteen=dfs[5]
year_twelve=dfs[6]
year_eleven=dfs[7]
year_ten=dfs[8]
year_nine=dfs[9]
year_eight=dfs[10]
```
